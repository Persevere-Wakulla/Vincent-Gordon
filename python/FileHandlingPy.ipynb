Create and Open a File:

        Syntax:
                file_object = open(file_name, 'mode')
        
        The mode in the open function syntax will tell Python what operation we want to do on the file.

        For text file:
            'r' - Read Mode: Read mode is used only to read data from the file.
            'w' - Write Mode: This mode is used when you want to write data into the file or modify it. Remember write mode overwrites the present data in the file.
            'a' - Append Mode: Append mode is used to append data to the file. Remember data will be appended at the end of the file pointer.
            'r+' - Read or Write Mode: This mode is used when we want to write or read the data from the same file.
        
        Note: The above mentioned modes are for opening reading or writing text files only.
        
        While using Binary files. we have to use the same modes with the letter 'b' at the end. So that Python can understand that we are interacting with binary files.
                'wb' - Open a file for write only mode in the binary format.
                'rb' - Open a file for the read-only mode in the binary format.
                'ab' - Open a file for appending only mode in the binary format.
                'rb+' - Open a file for the read and write only mode in the binary format.
                'ab+' - Open a file for appending and read-only mode in the binary format.

Image File Handling:

```python
from PIL import Image,ImageDraw,ImageFont
from IPython.display import display

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
location1 = r'./Images/TcDivision.jpg'
location2 = r'./Images/xbox.png'
```

```python
im1 = Image.open(location1)
im2 = Image.open(location2)
```

```python
im1
```

Properties of the Image:
                The format attribute identifies the source of an image. If the image was not read from a file, it is set to none

                The size attribute is a 2-tuple containing width and height (in pixels).
                
                The mode attribute defines the number and names of the bands in the image and also the pixel type and depth Common modes are:

                        'L' (luminance) for greyscale images.
                        'RGB' for true color images.
                        'CMYK' for pre press images.

```python
im1.format, im1.size,im1.mode
```

```python
im2.format, im2.size, im2.mode
```

```python
plt.figure(figsize=(12,5))

plt.subplot(121)
plt.imshow(im1)

plt.subplot(122)
plt.imshow(im2)
```

Cropping:
        box = (100,100,400,400) 
        region = im1.crop(box)

        The region is defined by a 4-tuple, Where coordinates are (left, upper, right, lower).

        The Python imaging Library uses a coordinate system with (0,0) in the upper left corner.

        Also note that coordinates refer to positions between the pixels, so the region in the above example is exactly 300 x 300 pixels.


```python
box = (150,20,300,500)
region = im1.crop(box)
```

```python
plt.imshow(region)
```

Transposing An Image:

```python
out = im1.transpose(Image.FLIP_LEFT_RIGHT)
out = im1.transpose(Image.FLIP_TOP_BOTTOM)
out = im1.transpose(Image.ROTATE_90)
out = im1.transpose(Image.ROTATE_180)
out = im1.transpose(Image.ROTATE_270)
plt.imshow(out)
```
